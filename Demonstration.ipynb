```python
import sqlite3

from google_keyword_graph.get_suggestions import store_suggestions, create_table
from google_keyword_graph.create_graph import gen_graph

import networkx as nx
from IPython.display import HTML
```

```python
conn = sqlite3.connect('keywords.db')
cursor = conn.cursor()
```

# Create graph for keyword `python`

```python
keyword = 'python'
depth = 4
```

```python
create_table()
store_suggestions(keyword, depth)
```

```python
G = gen_graph(keyword)
```

```python
G.nodes
```

```python
G.edges
```

## Graph created by mathplotlib

```python
nx.draw(G, with_labels=True, font_size=12, edge_color='gray')
```
